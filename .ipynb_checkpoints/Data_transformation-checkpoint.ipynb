```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import closed deals csv
deals_file = "Resources/closed_deals.csv"
deals_df= pd.read_csv(deals_file)
deals_df.head()
```

```python
#import reviews csv
reviews_file = "Resources/order_reviews.csv"
reviews_df= pd.read_csv(reviews_file)
reviews_df.head()
```

```python
#import orders csv
order_items_file = "Resources/order_items.csv"
order_items_df= pd.read_csv(order_items_file)
order_items_df.head()
```

```python
#join review df with orders df
reviews_joined =reviews_df.merge(order_items_df, on = 'order_id')
reviews_joined.head()
```

```python
#clean up joined dataframe
reviews_joined = reviews_joined[['review_id','order_id', 'review_score', 'seller_id', 'price']].copy()
reviews = reviews_joined.drop_duplicates('review_id')
reviews = reviews.set_index('review_id')

reviews
```

```python
#clean up closed deals df
sellers_df = deals_df[['seller_id', 'business_segment']]
sellers_df.drop_duplicates('seller_id', inplace=True)
sellers =sellers_df.set_index('seller_id')
```

```python
#establish connection
rds_connection_string = "postgres:postgres@localhost:5432/olist"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#export reviews df to sql table
reviews.to_sql(name='reviews', con=engine, if_exists='append', index=True)
```

```python
#export sellers df to sql table
sellers.to_sql(name='deals', con=engine, if_exists='append', index=True)
```
